```python
data.rename(columns= {'Cust_name':'Customr_name'})
```

```python
data[['Age', 'Orders', 'Amount']].describe()
```

```python
ax = sns.countplot(x = 'Gender',data = data,color='green') # x count value will be represented by y asxi

for bars in ax.containers:# used to represent the total values of male anf female
    ax.bar_label(bars)
```

```python
plt.show()
```

```python
sales_gen = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)#as_index=fase means not to group by columns
```

```python
sales_gen
```

```python
sales_gen = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,color='purple')
plt.show()
```

```python
sales_gen = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(6, 6))  # Optional: Set figure size
plt.pie(sales_gen['Amount'], labels=sales_gen['Gender'], autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Gender')
plt.axis('equal')  # Make it a circle
plt.show()
```

```python
ax = sns.countplot(data = data, x = 'Age Group', hue = 'Gender', palette=['red', 'green']) #hue is foe fender classification

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_age = data.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age,palette=['red', 'green', 'blue', 'orange'])
plt.show()
```

```python
sales_state = data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)},)# rcs runtime configuration and used for customisation n graph 
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
plt.show()
```

```python
sales_state = data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
plt.show()
```

```python
sales_state = data.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.show()
```

```python
sns.set(rc={'figure.figsize': (20, 5)})

# Use a color palette with enough distinct colors
ax = sns.countplot(data=data, x='Occupation', palette='Set3')  # You can also try 'husl', 'pastel', 'Set2', etc.

# Add value labels on top of each bar
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Count of People by Occupation")
plt.xticks(rotation=45)  # Rotate labels if needed
plt.show()
```

```python
sales_state = data.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',palette='Set3')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = data, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
plt.show()
```

```python
sales_state = data.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
plt.show()
```
